```python
import sympy as sp
import numpy as np
```

**Taller 1**
Ejercicio 5 de la sección 1.1.6

```python
NumMasas = int(input("Ingrese el numero de masas:"))
Masas = []
Posiciones = []
dim = int(input("Ingrese la dimensión del espacio con un número entero:"))



for i in range(NumMasas):
  m_i = sp.sympify(input("Ingrese las masas separadas por un enter:"))
  Masas.append(m_i)

#Apartado donde pido las masas, así no aparece intercalado
for i in range(NumMasas):
  ent = input("Ingrese las coordenadas de las masas separadas por una coma:")
  coors = ent.split(",")
  if len(coors) != dim:
    raise ValueError("El número de coordenadas no es válido para la dimensión:")

  coor = tuple(sp.sympify(c) for c in coors)
  Posiciones.append(coor)

#Fórmula para encontrar el centro de masa del sistema dado
Masa_Total = sum(Masas)
r_cm = []

for d in range(dim):
  numerador = sum(Masas[i]*Posiciones[i][d] for i in range(NumMasas))
  r_cm.append(sp.sympify(numerador/Masa_Total))
print ("El centro de masa se ubica en:", tuple(r_cm))
```

Ejercicios 6 de la sección 1.2.7

```python
n = int(input("Ingrese la dimensión del espacio con un número entero: "))


#Aquí se pide que se ingresen los componentes de los vectores

   #el comando sp.Matrix es para convertir la entrada en una matriz, que es como sympy reconoce a los vectores
   #el float es para tratar ese tipo de datos, que pueden ser no enteros; y el i+1 determina que los vectores comiencen
   # a enumerarse desde el indice 1 y no desde el 0.
   #Por otro lado, el for i in range indica cuantas componentes debe tener el vector de acuerdo a la dimensión, lo que
   # se repita el comando tantas veces se necesite.


print("Ingrese las coordenadas del vector a separadas por un enter:")
a = sp.Matrix([float(input(f"a_{i+1} = ")) for i in range(n)])

print("Ingrese las coordenadas del vector b separadas por un enter:")
b = sp.Matrix([float(input(f"b_{i+1} = ")) for i in range(n)])

print("Ingrese las coordenadas del vector c separadas por un enter:")
c = sp.Matrix([float(input(f"c_{i+1} = ")) for i in range(n)])

print("Ingrese las coordenadas del vector d separadas por un enter :")
d = sp.Matrix([float(input(f"d_{i+1} = ")) for i in range(n)])


#Las operaciones de las combinaciones lineales

print("\n--- Combinaciones lineales ---")
print("a + b + c + d =", a + b + c + d)
print("a + b - c - d =", a + b - c - d)
print("a - b + c - d =", a - b + c - d)
print("-a + b - c + d =", -a + b - c + d)


#Angulo entre a y los vectores

print("\n--- Ángulo entre a y los vectores base ---")

# Matriz identidad: de aquí salen los vectores base
bases = [sp.eye(n).col(i) for i in range(n)]

for i, ei in enumerate(bases):
    angulo = sp.acos(a.dot(ei) / (sp.sqrt(a.dot(a)) * sp.sqrt(ei.dot(ei))))
    print(f"Ángulo entre a y e_{i+1} =", angulo)


#Norma de los vectores

print("\n---- Magnitudes ---")
print("|a| =", sp.sqrt(a.dot(a)))
print("|b| =", sp.sqrt(b.dot(b)))
print("|c| =", sp.sqrt(c.dot(c)))
print("|d| =", sp.sqrt(d.dot(d)))


#Angulos entre los vectores

print("\n---- Ángulos entre vectores ---")

angulo_ab = sp.acos(a.dot(b) / (sp.sqrt(a.dot(a)) * sp.sqrt(b.dot(b))))
angulo_cd = sp.acos(c.dot(d) / (sp.sqrt(c.dot(c)) * sp.sqrt(d.dot(d))))

print("Ángulo entre a y b =", angulo_ab)
print("Ángulo entre c y d =", angulo_cd)


#Proyección de a sobre b

print("\n--- Proyección de a sobre b ---")

proy_a_b = (a.dot(b) / b.dot(b)) * b
print("Proyección de a sobre b =", proy_a_b)



# Coplanaridad (solo en R³)

print("\n---  Coplanaridad ---")

if n == 3:
    triple = a.dot(b.cross(c))
    print("a · (b × c) =", triple)
else:
    print("La coplanaridad con producto cruz solo se define en R³")



#Producto punto (a + b) · (c + d)

print("\n---  Producto punto ---")
print("(a + b) · (c + d) =", (a + b).dot(c + d))


# 10. (h) Productos cruz y ángulos

print("\n---  Productos cruz ---")

if n == 3:
    axb = a.cross(b)
    bxc = b.cross(c)
    cxd = c.cross(d)

    print("a × b =", axb)
    print("b × c =", bxc)
    print("c × d =", cxd)

    angulo_axb_d = sp.acos(axb.dot(d) / (sp.sqrt(axb.dot(axb)) * sp.sqrt(d.dot(d))))
    print("Ángulo entre (a × b) y d =", angulo_axb_d)
else:
    print("El producto cruz solo existe en R³")



#Producto triple escalar

print("\n--- 1 Producto triple escalar ---")

if n == 3:
    print("c · (a × b) =", c.dot(a.cross(b)))
```
